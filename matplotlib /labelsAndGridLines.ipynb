## Labels and Title Customization Notes

### What is Label and Title Customization?

**Label and Title Customization** refers to the process of enhancing the readability and visual appeal of plots by customizing their labels, title, tick parameters, and other text elements.

### Key Rules

1. **Customizing Titles**
   - Set the title using `plt.title()` with parameters like `fontsize`, `family`, `fontweight`, and `color`.
2. **Customizing Axis Labels**
   - Customize x-axis and y-axis labels using `plt.xlabel()` and `plt.ylabel()` with similar parameters.
3. **Adjusting Tick Parameters**
   - Modify tick parameters globally using `plt.tick_params()` to change the color of ticks.
4. **Aligning Text Elements**
   - Ensure text elements like titles, labels, and ticks are aligned properly for better visual impact.

### Examples

1. **Customizing Titles**

    ```python
    plt.title("Class size", fontsize=20,
                        family = "Arial",
                        fontweight = "bold",
                        color = "#fc1c1c")
    ```

2. **Customizing Axis Labels**

    ```python
    plt.xlabel("Year", fontsize=20,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#ec4f4f")

    plt.ylabel("Students", fontsize=20,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#ec4f4f")
    ```

3. **Adjusting Tick Parameters**

    ```python
    plt.tick_params(axis="both",
                colors = "#ec4f4f")
    ```

4. **Aligning Text Elements**

    ```python
    plt.xticks(x)
    ```

# Grid Lines Customization Notes

### What is Grid Line Customization?

**Grid Line Customization** refers to the process of adding and customizing grid lines in plots to improve readability and visual appeal. Grid lines help in identifying specific values on the axes more easily.

### Key Rules

1. **Adding Grid Lines**
   - Enable grid lines using `plt.grid()` with parameters like `axis` (e.g., "x", "y", or both) and `linestyle`.
2. **Customizing Grid Line Appearance**
   - Customize the appearance of grid lines with parameters like `color`, `linewidth`, and `alpha`.
3. **Controlling Grid Visibility**
   - Hide or show grid lines using the `visible` parameter.
4. **Adjusting Grid Line Spacing**
   - Adjust the spacing between grid lines using the `which` parameter (e.g., "major" or "minor").

### Examples

1. **Adding Grid Lines**

    ```python
    plt.grid(axis="y", linestyle="--")
    ```

2. **Customizing Grid Line Appearance**

    ```python
    plt.grid(axis="x", color="lightblue", linewidth=0.5, alpha=0.7)
    ```
3. **Controlling Grid Visibility**

    ```python
    plt.grid(axis="y", visible=False)
    ```

4. **Adjusting Grid Line Spacing**

    ```python
    plt.minorticks_on()
    plt.grid(which="minor", linestyle=":", linewidth=0.5, alpha=0.3)
    ```

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([5, 32, 48, 10])
y3 = np.array([20, 30, 45, 40])

plt.grid(axis = "y",
         linewidth = 2,
         color = "lightblue",
         linestyle = "dashed",
         alpha = 0.7)

plt.title("Class size", fontsize=20,
                        family = "Arial",
                        fontweight = "bold",
                        color = "#fc1c1c")

plt.xlabel("Year", fontsize=20,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#ec4f4f")

plt.ylabel("Students", fontsize=20,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#ec4f4f")

plt.tick_params(axis="both",
                colors = "#ec4f4f")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x)

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 10, 15, 20, 25])

plt.grid(axis="both", linestyle="-.")
plt.grid(axis="x", color="#ec4f4f", linewidth=1.3, alpha=0.7)
plt.minorticks_on()
plt.grid(which="minor", linestyle=":", linewidth=0.5, alpha=0.3)

plt.plot(x, y)
plt.show()
````
