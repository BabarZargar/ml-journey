## Labels and Title Customization Notes

### What is Label and Title Customization?

**Label and Title Customization** refers to the process of enhancing the readability and visual appeal of plots by customizing their labels, title, tick parameters, and other text elements.

### Key Rules

1. **Customizing Titles**
   - Set the title using `plt.title()` with parameters like `fontsize`, `family`, `fontweight`, and `color`.
2. **Customizing Axis Labels**
   - Customize x-axis and y-axis labels using `plt.xlabel()` and `plt.ylabel()` with similar parameters.
3. **Adjusting Tick Parameters**
   - Modify tick parameters globally using `plt.tick_params()` to change the color of ticks.
4. **Aligning Text Elements**
   - Ensure text elements like titles, labels, and ticks are aligned properly for better visual impact.

### Examples

1. **Customizing Titles**

    ```python
    plt.title("Class size", fontsize=20,
                        family = "Arial",
                        fontweight = "bold",
                        color = "#fc1c1c")
    ```

2. **Customizing Axis Labels**

    ```python
    plt.xlabel("Year", fontsize=20,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#ec4f4f")

    plt.ylabel("Students", fontsize=20,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#ec4f4f")
    ```

3. **Adjusting Tick Parameters**

    ```python
    plt.tick_params(axis="both",
                colors = "#ec4f4f")
    ```

4. **Aligning Text Elements**

    ```python
    plt.xticks(x)
    ```

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([5, 32, 48, 10])
y3 = np.array([20, 30, 45, 40])

plt.title("Class size", fontsize=20,
                        family = "Arial",
                        fontweight = "bold",
                        color = "#fc1c1c")

plt.xlabel("Year", fontsize=20,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#ec4f4f")

plt.ylabel("Students", fontsize=20,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#ec4f4f")

plt.tick_params(axis="both",
                colors = "#ec4f4f")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x)

plt.show()
````
