## Bar Charts

### What is a Bar Chart?

**Bar Chart** refers to a graphical representation of data where categories are compared by their heights. Each bar represents a category, and the length of the bar indicates the value associated with that category.

### Key Rules

1. **Defining Categories and Values**
   - Use arrays or lists to define categories and their corresponding values.
2. **Creating a Vertical Bar Chart**
   - Use `plt.bar()` to create a vertical bar chart by passing categories and values.
3. **Customizing Colors**
   - Set the color of bars using the `color` parameter.
4. **Adding Labels and Titles**
   - Add labels for the x-axis, y-axis, and title using `plt.xlabel()`, `plt.ylabel()`, and `plt.title()`.

### Examples

1. **Defining Categories and Values**

    ```python
    categories = np.array(["Grains", "Fruit", "Vegetables", "Protein", "Dairy", "Sweets"])
    values = np.array([4, 3, 2, 5, 3, 1])
    ```

2. **Creating a Vertical Bar Chart**

    ```python
    plt.bar(categories, values, color="skyblue")
    ```

3. **Customizing Colors**

    ```python
    # For horizontal bar chart
    # plt.barh(categories, values, color="skyblue")
    ```

4. **Adding Labels and Titles**

    ```python
    plt.title("daily Consumption")
    plt.xlabel("Food")
    plt.ylabel("Quantity")
    ```

## Pie Charts

### What is a Pie Chart?

**Pie Chart** refers to a circular chart divided into slices, each representing a proportion of the whole. It is ideal for visualizing distribution among categories.

### Key Rules

1. **Defining Categories and Values**
   - Use arrays or lists to define categories and their corresponding values.
2. **Creating a Pie Chart**
   - Use `plt.pie()` to create a pie chart by passing values, labels, colors, and other optional parameters.
3. **Customizing Colors**
   - Set the color of slices using the `colors` parameter.
4. **Exploding Slices**
   - Explode one or more slices from the center using the `explode` parameter.
5. **Adding Labels and Titles**
   - Add labels for the chart and title using `plt.title()`.

### Examples

1. **Defining Categories and Values**

    ```python
    categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
    values = np.array([350, 250, 275, 200])
    colors = ["red", "yellow", "blue", "green"]
    ```

2. **Creating a Pie Chart**

    ```python
    plt.pie(values, labels=categories,
                    autopct="%1.1f%%",
                    colors=colors,
                    explode=[0, 0, 0, 0.1],
                    shadow=True,
                    startangle=90)
    ```

````python
import matplotlib.pyplot as plt
import numpy as np

# Bar chart = compare categories of data by representing each category with a bar chart

categories = np.array(["Grains", "Fruit", "Vegetables", "Protein", "Dairy", "Sweets"])

values = np.array([4, 3, 2, 5, 3, 1])

plt.bar(categories, values, color = "skyblue")

#for a horizontal bar chart
#plt.barh(categories, values, color = "skyblue") 

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

#Pie chart = Circular chart divided into slices to show percentages of the total.
#Good for visualizing distribution among categories.

categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([350, 250, 275, 200])
colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels=categories,
                autopct="%1.1f%%",
                colors=colors,
                explode=[0, 0, 0, 0.1],
                shadow=True,
                startangle=90)

plt.title("NIT Srinagar")

plt.show()
````
