## Plot Customization Notes

### What is Plot Customization?

**Plot Customization** refers to the process of adjusting various aspects of a plot to enhance its appearance and make it more informative and visually appealing. This includes setting colors, line styles, markers, and labels.

### Key Rules

1. **Setting Line Styles**
   - Use the `linestyle` parameter to define the style of the lines.
2. **Customizing Markers**
   - Customize markers with the `marker`, `markersize`, `markerfacecolor`, and `markeredgecolor` parameters.
3. **Changing Colors**
   - Set colors for lines and other plot elements using the `color` parameter.
4. **Adding Labels and Titles**
   - Use `xlabel`, `ylabel`, and `title` to add labels and a title to your plot.

### Examples

1. **Setting Line Styles**

    ```python
    line_style = dict(marker=".",
                      markersize=15,
                      markerfacecolor="#fc1c1c",
                      markeredgecolor="#321cfc",
                      linestyle="solid",
                      linewidth=4)
    ```

2. **Customizing Markers**

    ```python
    plt.plot(x, y1, color="#1c9bfc", **line_style)
    ```

3. **Changing Colors**

    ```python
    plt.plot(x, y2, color="#27fc1c", **line_style)
    ```

4. **Adding Labels and Titles**

    ```python
    plt.xlabel("Year")
    plt.ylabel("Value")
    plt.title("Customized Plot")
    ```

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([5, 32, 48, 10])
y3 = np.array([20, 30, 45, 40])

line_style = dict(marker=".",
                  markersize=15,
                  markerfacecolor="#fc1c1c",
                  markeredgecolor = "#321cfc",
                  linestyle="solid",
                  linewidth=4)

plt.plot(x, y1, color="#1c9bfc", **line_style)
plt.plot(x, y2, color="#27fc1c",  **line_style)
plt.plot(x, y3, color="#e61cfc", **line_style)


plt.show()
````
