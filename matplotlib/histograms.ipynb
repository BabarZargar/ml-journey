## Histograms Notes

### What is a Histogram?

**Histogram** refers to a graphical representation of the distribution of quantitative data. It groups values into bins (intervals) and counts how many fall in each range, allowing you to visualize the frequency distribution.

### Key Rules

1. **Defining Data**
   - Use an array or list to define the data points.
2. **Creating a Histogram**
   - Use `plt.hist()` to create a histogram by passing the data along with other optional parameters like number of bins (`bins`), color, and edgecolor.
3. **Customizing Colors**
   - Set the color of the bars using the `color` parameter.
4. **Adjusting Edge Color**
   - Change the edge color of the bins using the `edgecolor` parameter.
5. **Adding Labels and Titles**
   - Add labels for the x-axis, y-axis, and title using `plt.xlabel()`, `plt.ylabel()`, and `plt.title()`.

### Examples

1. **Defining Data**

    ```python
    scores = np.random.normal(loc=80, scale=10, size=100)
    scores = np.rint(scores).astype(int)        # round to nearest integer
    scores = np.clip(scores, 0, 100)            # keep within 0-100
    ```

2. **Creating a Histogram**

    ```python
    plt.hist(scores, bins=10,
                     color="lightgreen",
                     edgecolor="black")
    ```

````python
import matplotlib.pyplot as plt
import numpy as np

# Histogram = A visual representation of the distribution of quantitative data.
# Make scores integer and annotate each bar with its count.

scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.rint(scores).astype(int)        # round to nearest integer
scores = np.clip(scores, 0, 100)            # keep within 0-100

plt.hist(scores, bins=10,
                 color="lightgreen",
                 edgecolor="black")

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("no. of students")

plt.show()
````
