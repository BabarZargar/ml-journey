## NumPy Arithmetic

### Scalar Arithmetic
A scalar is a single number. When you apply an operation to an array with a scalar,
it applies to **every element** automatically — no loops needed.

| Operation | Symbol | Example |
|-----------|--------|---------|
| Addition | `+` | `array + 1` |
| Subtraction | `-` | `array - 2` |
| Multiplication | `*` | `array * 3` |
| Division | `/` | `array / 4` |
| Power | `**` | `array ** 5` |


````python
import numpy as np

# Scaler arithmetic

array = np.array([1, 2, 3])
print(array + 1)
print(array - 2)
print(array * 3)
print(array / 4)
print(array ** 5)
````

### Vectorized Math Functions
NumPy has built-in math functions that work on entire arrays at once.

| Function | Meaning |
|----------|---------|
| `np.sqrt(array)` | square root of each element |
| `np.round(array)` | round each element |
| `np.floor(array)` | round down each element |
| `np.ceil(array)` | round up each element |
| `np.pi` | returns the value of π (3.14159...) |

````python
import numpy as np 

# Vectorized math funcs

array = np.array([1.01, 2.5, 3.99])
print(np.sqrt(array))
print(np.round(array))
print(np.floor(array))
print(np.ceil(array))
print(np.pi)
````

### Element-wise Arithmetic
When you do arithmetic between two arrays of the **same size**,
NumPy operates on matching elements — index 0 with index 0, index 1 with index 1, and so on.

| Operation | Example |
|-----------|---------|
| Addition | `array1 + array2` |
| Subtraction | `array2 - array1` |
| Multiplication | `array1 * array2` |
| Division | `array1 / array2` |
| Power | `array2 ** array1` |

````python
import numpy as np 

# Element wise arithmetic

array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

print(array1 + array2)
print(array2 - array1)
print(array1 * array2)
print(array1 / array2)
print(array2 ** array1)
````

### Comparison Operators
You can compare an array against a value — returns a **boolean array** of True/False.

```python
scores == 100   # which elements equal 100?
scores >= 70    # which elements are passing?
```

You can also use the boolean result to **modify** elements:
```python
scores[scores < 70] = 0   # set all failing scores to 0
```
This is called **boolean indexing** — very powerful and used a lot in ML.

````python
import numpy as np 

# Comparison operators

scores = np.array([91, 55, 100, 73, 82, 64])

print(scores == 100) #Check if anyone got 100
print(scores >= 70) #Check passsing students

scores[scores < 70] = 0
print(scores)
````
