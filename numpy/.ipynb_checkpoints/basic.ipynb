## NumPy Basics

### What is NumPy?
NumPy (Numerical Python) is a library for working with arrays and mathematical operations.
It is faster and more efficient than regular Python lists.

### Importing NumPy
Always import NumPy as `np` — this is the universal convention.
`import numpy as np`

### Creating an Array
Use `np.array()` to create an array from a Python list.
`array = np.array([1, 2, 3, 4])`

### Type of a NumPy array
The type of a NumPy array is `numpy.ndarray` (n-dimensional array).

### Array Arithmetic
Unlike Python lists, you can do math directly on arrays.
- `array * 2` multiplies every element by 2
- No need to loop through elements — NumPy handles it automatically
- This is called **vectorized operation**

```python
import numpy as np
print(np.__version__)
```

```python
import numpy as np
array = np.array([1, 2, 3, 4])
array = array*2
print(array)
print(type(array))
```
