## NumPy Broadcasting

### What is Broadcasting?
Broadcasting is NumPy's way of performing arithmetic on arrays of **different shapes**.
Instead of requiring both arrays to be the same size, NumPy automatically "stretches"
the smaller array to match the larger one.

### How it Works
NumPy compares the shapes of two arrays from right to left.
A dimension is compatible if:
- They are equal, **or**
- One of them is 1

### Example 1 — (1,4) × (4,1)
```python
array1.shape = (1, 4)  # 1 row, 4 columns  → [1, 2, 3, 4]
array2.shape = (4, 1)  # 4 rows, 1 column  → [[1],[2],[3],[4]]
```
Result shape: **(4, 4)** — each element of array1 multiplied by each element of array2.
Produces a multiplication table.

### Example 2 — (4,4) × (4,1)
```python
array1.shape = (4, 4)  # full 4x4 grid
array2.shape = (4, 1)  # single column
```
Result shape: **(4, 4)** — each row of array1 gets multiplied by the matching scalar in array2.

### Example 3 — (1,10) × (10,1)
```python
array1.shape = (1, 10)   # 1 row, 10 columns
array2.shape = (10, 1)   # 10 rows, 1 column
```
Result shape: **(10, 10)** — produces a full 10×10 multiplication table.

### Key Rule
| array1 shape | array2 shape | result shape |
|-------------|-------------|-------------|
| (1, 4) | (4, 1) | (4, 4) |
| (4, 4) | (4, 1) | (4, 4) |
| (1, 10) | (10, 1) | (10, 10) |

### Why Broadcasting Matters in ML
Broadcasting is used constantly in ML for:
- Scaling features in a dataset
- Computing distances between points
- Applying weights to neural network layers

No loops needed — NumPy handles it all automatically and much faster.

````python
import numpy as np

array1 = np.array([[1, 2, 3, 4]])
array2 = np.array([[1], [2], [3], [4]])

print(array1.shape)
print(array2.shape)

print(array1*array2)
````
