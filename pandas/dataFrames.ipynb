
## Pandas DataFrame

### What is a DataFrame?

A **DataFrame** is a two-dimensional labeled data structure with columns and rows. Think of it as a table or a spreadsheet.


### Creating a DataFrame

| Code | Meaning |
|------|---------|
| `pd.DataFrame(data)` | create a DataFrame from a dictionary |


### Custom Index

| Code | Meaning |
|------|---------|
| `pd.DataFrame(data, index=[...])` | assign custom row labels |
| `df.loc["Student 2"]` | access a row by **label** |
| `df.iloc[3]` | access a row by **integer position** |


### Adding a New Column

| Code | Meaning |
|------|---------|
| `df["Branch"] = [...]` | add a new column with a list of values |



### Adding a New Row

| Code | Meaning |
|------|---------|
| `pd.DataFrame([{...}], index=[...])` | create a single-row DataFrame |
| `pd.concat([df, new_row])` | append the new row to the original |


### Adding Multiple Rows

| Code | Meaning |
|------|---------|
| `pd.DataFrame([{...}, {...}], index=[...])` | create a multi-row DataFrame |
| `pd.concat([df, new_rows])` | append multiple rows at once |



### Key Rules

1. **Dictionary keys become column names** — values must be lists of equal length.
2. **Default index is 0, 1, 2...** — use `index=` to set custom row labels.
3. **`.loc[]` uses labels** — access rows by their custom index name.
4. **`.iloc[]` uses position** — access rows by integer position (0-based).
5. **Add columns with `df["Col"] = [...]`** — the list length must match the number of rows.
6. **Add rows with `pd.concat()`** — create a new DataFrame for the row(s), then concatenate.
7. **`pd.concat()` does not modify in place** — you must reassign: `df = pd.concat([df, new_row])`.

````python
import pandas as pd

data = {"Name": ["Babar", "Hamzah", "Natiq", "Basit"],
        "Age": [18, 18, 17, 19]
        }

df = pd.DataFrame(data)

print(df)
````

````python
import pandas as pd

data = {"Name": ["Babar", "Hamzah", "Natiq", "Basit"],
        "Age": [18, 18, 17, 19]
        }

df = pd.DataFrame(data, index=["Student 1","Student 2", "Student 3", "Student 4"])

print(df)
print(df.loc["Student 2"])
print(df.iloc[3])
````

````python
import pandas as pd

data = {"Name": ["Babar", "Hamzah", "Natiq", "Basit"],
        "Age": [18, 18, 17, 19]
        }

df = pd.DataFrame(data, index=["Student 1","Student 2", "Student 3", "Student 4"])

#add a new column
df ["Branch"] = ["ECE", "N/A", "IT", "CS"]

#add a new row
new_row = pd.DataFrame([{"Name": "Dayan", "Age": 19, "Branch": "CIV"}],
                       index = ["Student 5"])
df = pd.concat([df, new_row])

print(df)
````

````python
import pandas as pd

data = {"Name": ["Babar", "Hamzah", "Natiq", "Basit"],
        "Age": [18, 18, 17, 19]
        }

df = pd.DataFrame(data, index=["Student 1","Student 2", "Student 3", "Student 4"])

#add a new column
df ["Branch"] = ["ECE", "N/A", "IT", "CS"]

#add a new row
new_rows = pd.DataFrame([{"Name": "Dayan", "Age": 19, "Branch": "CIV"},
                         {"Name": "Muhaimin", "Age": 20, "Branch": "CS"}],
                       index = ["Student 5", "Student 6"])
df = pd.concat([df, new_rows])

print(df)
````
