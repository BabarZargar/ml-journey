## Pandas Aggregation

### What is Aggregation?

**Aggregation** refers to applying statistical functions to a DataFrame or a Series. This allows you to summarize data and extract meaningful insights.

### Key Rules

1. **Grouping data by one or more columns** — use the `groupby()` method on a DataFrame.
2. **Applying aggregation functions** — such as `mean()`, `sum()`, `min()`, `max()`, `count()`, etc., to grouped data.
3. **Combining multiple aggregation functions** — use dictionaries in the `agg()` method to apply different functions to different columns.
4. **Filtering groups based on conditions** — use boolean indexing or other methods to filter groups after aggregation.

### Examples

1. **Grouping by a Single Column and Applying Mean**

    ```python
    group = df.groupby("Type 1")
    print("Attack\n", group["Attack"].mean())
    ```

2. **Grouping by a Single Column and Applying Sum**

    ```python
    print("\nDefense\n", group["Defense"].sum())
    ```

3. **Grouping by a Single Column and Applying Min**

    ```python
    print("\nSpeed\n", group["Speed"].min())
    ```

4. **Counting the Number of Rows in Each Group**

    ```python
    print(group["Attack"].count())
    ```

5. **Applying Multiple Aggregation Functions to Different Columns**

    ```python
    agg_results = group.agg({"Attack": ["mean", "sum"], "Defense": "max"})
    print(agg_results)
    ```

6. **Filtering Groups Based on Aggregated Results**

    ```python
    filtered_groups = group[group["Attack"].mean() > 100]
    print(filtered_groups)
    ```

7. **Using `transform()` for Applying Functions and Storing Results in a New Column**

    ```python
    df["Attack_Rank"] = group["Attack"].rank()
    print(df)
    ```

````python
import pandas as pd

# Aggregate functions 
# Reduces a set of values into a single summary value 
# Used to summarize and analyze data 
# Often used with the groupby() function


df = pd.read_csv("Pokemon.csv")

#Whole dataframe
print("Mean: \n", df.mean(numeric_only=True))
print("\nSum: \n", df.sum(numeric_only=True))
print("\nMin: \n", df.min(numeric_only=True))
print("\nMax: \n", df.max(numeric_only=True))
print("\nCount: ", df.count())

````

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#Single column
print("\nMean of attack: \n", df["Attack"].mean())
print("\nSum of defense: \n", df["Defense"].sum())
print("\nMin of speed: \n", df["Speed"].min())
print("\nMax of speeed: \n", df["Speed"].max())
print("\nCount:", df["Type 2"].count())
````

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

group = df.groupby("Type 1")

print("Attack\n", group["Attack"].mean())
print("\nDefense\n", group["Defense"].sum())
print("\nSpeed\n", group["Speed"].min())
print(group["Attack"].count())
````
