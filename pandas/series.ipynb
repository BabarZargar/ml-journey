## Pandas Series

### What is a Series?

A **Series** is a one-dimensional labeled array. It is like a NumPy array with an **index** attached to each value.

```python
import pandas as pd
```

### Creating a Series

| Code | Meaning |
|------|---------|
| `pd.Series(data)` | create a Series from a list |
| `pd.Series(data, index=["a", "b", "c"])` | create a Series with custom index labels |
| `pd.Series(dictionary)` | create a Series from a dictionary (keys become index) |

```python
data1 = [100, 102, 104]
data2 = [101.1, 103.3, 105.5]
data3 = ["a", "b", "c"]

series1 = pd.Series(data1)   # dtype: int64
series2 = pd.Series(data2)   # dtype: float64
series3 = pd.Series(data3)   # dtype: object
```

&gt; **Output:**
&gt; ```
&gt; 0    100
&gt; 1    102
&gt; 2    104
&gt; dtype: int64
&gt; ```

### Custom Index

| Code | Meaning |
|------|---------|
| `pd.Series(data, index=["a", "b", "c"])` | assign custom labels to each element |
| `series.loc["a"]` | access value by **label** |
| `series.iloc[0]` | access value by **integer position** |

```python
data = [100, 102, 104]
series = pd.Series(data, index=["a", "b", "c"])

series.loc["c"] = 200        # change value at label "c"

print(series)
print(series.loc["a"])       # 100  (by label)
print(series.iloc[0])         # 100  (by position)
```

&gt; **Output:**
&gt; ```
&gt; a    100
&gt; b    102
&gt; c    200
&gt; dtype: int64
&gt; 100
&gt; 100
&gt; ```

### Boolean Filtering

| Code | Meaning |
|------|---------|
| `series[series &gt;= 200]` | all elements greater than or equal to 200 |
| `series[series &lt; 200]` | all elements less than 200 |

```python
data = [100, 102, 104, 200, 202]
series = pd.Series(data, index=["a", "b", "c", "d", "e"])

print(series[series &gt;= 200])   # d: 200, e: 202
print(series[series &lt; 200])    # a: 100, b: 102, c: 104
```

&gt; **Output:**
&gt; ```
&gt; d    200
&gt; e    202
&gt; dtype: int64
&gt;
&gt; a    100
&gt; b    102
&gt; c    104
&gt; dtype: int64
&gt; ```

### Series from a Dictionary

| Code | Meaning |
|------|---------|
| `pd.Series(dictionary)` | keys become index labels, values become data |
| `series.loc["Day 3"] += 50` | update a value by its label |
| `series.loc["Day 2"]` | access a single value by label |

```python
calories = {"Day 1": 2600, "Day 2": 2100, "Day 3": 2700}
series = pd.Series(calories)

series.loc["Day 3"] += 50     # Day 3 becomes 2750

print(series)
print(series.loc["Day 2"])    # 2100
print(series[series &gt; 2500])  # Day 1: 2600, Day 3: 2750
print(series[series &lt; 2500])  # Day 2: 2100
```

&gt; **Output:**
&gt; ```
&gt; Day 1    2600
&gt; Day 2    2100
&gt; Day 3    2750
&gt; dtype: int64
&gt; 2100
&gt; ```

### Key Rules

1. **A Series has an index** — every value has a label (default is 0, 1, 2...).
2. **`.loc[]` uses labels** — access by the custom index name.
3. **`.iloc[]` uses position** — access by integer position (0-based).
4. **Boolean filtering works like NumPy** — `series[series &gt; 100]` returns matching rows.
5. **Dictionaries auto-create index** — keys become the index when passed to `pd.Series()`.
```

````python
import pandas as pd

data1 = [100, 102, 104]
data2 = [101.1, 103.3, 105.5]
data3 = ['a', 'b', 'c']

series1 = pd.Series(data1)
series2 = pd.Series(data2)
series3 = pd.Series(data3)

print(series1)
print(series2)
print(series3)
````

````python
import pandas as pd

data = [100, 102, 104]

series = pd.Series(data, index=["a", "b", "c"])
series.loc["c"] = 200

print(series)
print(series.loc["a"])
print(series.iloc[0])
````
