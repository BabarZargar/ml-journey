```python
import pandas as pd

data1 = [100, 102, 104]
data2 = [101.1, 103.3, 105.5]
data3 = ['a', 'b', 'c']

series1 = pd.Series(data1)
series2 = pd.Series(data2)
series3 = pd.Series(data3)

print(series1)
print(series2)
print(series3)
```

```python
import pandas as pd

data = [100, 102, 104]

series = pd.Series(data, index=["a", "b", "c"])
series.loc["c"] = 200

print(series)
print(series.loc["a"])
print(series.iloc[0])
```

```python
import pandas as pd

data = [100, 102, 104, 200, 202]

series = pd.Series(data, index=["a", "b", "c", "d", "e"])

print(series[series >= 200])
print(series[series<200])
```

```python
import pandas as pd

calories = {"Day 1": 2600, "Day 2": 2100, "Day 3": 2700}

series = pd.Series(calories)

series.loc["Day 3"] += 50

print(series)
print(series.loc["Day 2"])

print(series[series > 2500])
print(series[series < 2500])

```
