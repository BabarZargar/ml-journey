## Data Cleaning Notes

### What is Data Cleaning?

**Data Cleaning** refers to the process of correcting or removing inaccurate, incomplete, or irrelevant data from a dataset. This allows you to ensure that your analysis is based on high-quality data.

### Key Rules

1. **Handling Missing Values**
   - Identify and address missing values using methods like filling with mean, median, mode, or dropping rows/columns.
2. **Removing Duplicates**
   - Use the `drop_duplicates()` method to eliminate duplicate entries in your dataset.
3. **Standardizing Data Formats**
   - Ensure consistency in data formats (e.g., dates, numbers) across all entries.
4. **Converting Datatypes**
   - Convert columns to appropriate datatypes using functions like `astype()`.
5. **Replacing data**
   - Fix inconsistent values

### Examples

1. **Handling Missing Values**

    ```python
    df = df.fillna({"Type 2": "None"})
    #or
    df = df.dropna(subset = ["Type 2"])
    ```

2. **Removing Duplicates**

    ```python
    df = df.drop_duplicates()  # Remove duplicate rows
    ```

3. **Standardizing Data Formats**

    ```python
    df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')  # Convert to datetime
    ```

4. **Converting Datatypes**

    ```python
    df["Speed"] = df["Speed"].astype(bool)  # Convert to integer type
    ```

5. **Replacing inconsistent values**

    ```python
    df["Type 1"] = df["Type 1"].replace(({"Grass": "GRASS",
                                     "Fire": "FIRE",
                                     "Water": "WATER"}))  #fix inconsistent values
    ```

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#drop irrelevant columns
df = df.drop(columns = ["Sp. Atk", "Sp. Def"])

print(df)
````

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#handle missing data
df = df.dropna(subset = ["Type 2"])

print(df.to_string())
````

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#handle missing data
df = df.fillna({"Type 2": "None"})

print(df.to_string())
````

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#fix inconsistent values
df["Type 1"] = df["Type 1"].replace({"Grass": "GRASS",
                                     "Fire": "FIRE",
                                     "Water": "WATER"})

print(df.to_string())
````

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#standardize text
df["Name"] = df["Name"].str.lower()

print(df.to_string())
````

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#fix data type
df["Speed"] = df["Speed"].astype(bool)

print(df.to_string())
````

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#remove duplicate values
df = df.drop_duplicates()

print(df.to_string())
````
