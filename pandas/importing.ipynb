## Pandas Importing

### What is Importing?

**Importing** refers to the process of bringing in data into a pandas DataFrame. This can be done using various functions depending on the file format.

### Reading Different File Formats

| Code | Meaning |
|------|---------|
| `pd.read_csv("file.csv")` | read data from a CSV file |
| `pd.read_json("file.json")` | read data from a JSON file |

### Key Rules

1. **Select the correct function based on the file format** — pandas provides specific functions like `read_csv`, `read_excel`, etc.
2. **File paths can be relative or absolute** — ensure the path points to the correct location of the file.
3. **Data formats are inferred by default** — pandas tries to guess the data type, but it's good practice to verify.

### Examples

1. **Reading a CSV File**

    ```python
    import pandas as pd

    df = pd.read_csv("pokemon.csv")
    print(df)
    ```

2. **Reading a JSON File**

    ```python
    import pandas as pd

    df = pd.read_json("data.json")
    print(df)
    ```

3. **Printing Entire DataFrame (for large datasets)**

    ```python
    print(df.to_string())
    ```

````python
import pandas as pd

df = pd.read_csv("pokemon.csv") #read_json for json file

print(df) #reads only starting and last few elements
````

````python
import pandas as pd

df = pd.read_csv("pokemon.csv") 

print(df.to_string()) #reads form start to end
````
