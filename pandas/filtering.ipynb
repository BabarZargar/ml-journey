## Pandas Filtering

### What is Filtering?

**Filtering** refers to selecting rows from a DataFrame based on specific conditions. This allows you to extract relevant data that meets certain criteria.

### Key Rules

1. **Use boolean indexing for filtering** — create a condition that returns a boolean Series, and use it to index the DataFrame.
2. **Combine multiple conditions using logical operators** — use `&` (and), `|` (or), and `~` (not) to combine conditions.
3. **Enclose conditions in parentheses** — to ensure correct operator precedence, use parentheses around each condition.
4. **Assign filtered data to a new DataFrame** — store the result of filtering in a new variable for further analysis or manipulation.

### Examples

1. **Filtering by a Single Condition**

    ```python
    strong_pokemon = df[df["Attack"] > 110]
    ```

2. **Filtering by Multiple Conditions with OR**

    ```python
    fire_type = df[(df["Type 1"] == "Fire") | 
                   (df["Type 2"] == "Fire")]
    ```

3. **Filtering by Multiple Conditions with AND and OR**

    ```python
    fire_flying = df[((df["Type 1"] == "Fire") &
                      (df["Type 2"] == "Flying")) |
                     ((df["Type 1"] == "Flying") &
                      (df["Type 2"] == "Fire"))]
    ```

4. **Handling Complex Conditions with Parentheses**

    ```python
    condition = ((df["Type 1"] == "Water") & 
                  (df["Attack"] > 80)) | 
                 ((df["Type 2"] == "Water") & 
                  (df["Defense"] > 70))
    water_strong_pokemon = df[condition]
    ```

5. **Negating Conditions with NOT**

    ```python
    non_flying = df[~(df["Type 1"] == "Flying")]
    ```

6. **Filtering by Multiple Columns**

    ```python
    condition = (df["Type 1"] == "Electric") & 
                 (df["HP"] > 90)
    electric_powerhouses = df[condition]
    ```

7. **Using Filtered Data for Further Analysis**

    ```python
    strong_fire_pokemon = df[(df["Attack"] > 120) & 
                               (df["Type 1"] == "Fire")]
    print(strong_fire_pokemon)
    ```

````python
import pandas as pd

df = pd.read_csv("Pokemon.csv")

#Filtering = keeping the rows that match a condition

strong_pokemon = df[df["Attack"] > 110]


fire_type = df[(df["Type 1"] == "Fire") | 
               (df["Type 2"] == "Fire")]

fire_flying = df[((df["Type 1"] == "Fire") &
                  (df["Type 2"] == "Flying")) |
                  ((df["Type 1"] == "Flying") &
                  (df["Type 2"] == "Fire"))]

print("strong pokemon:\n", strong_pokemon)
print("\nfire type:\n", fire_type)
print("\nfire and flying:\n", fire_flying)
````
